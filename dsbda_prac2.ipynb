```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/priya/Downloads/Student_information(Sheet1) (1).csv")
sorted_rscore= sorted(df['Reading_Score'])
```

```python
sorted_rscore
```

```python
q1 = np.percentile(sorted_rscore, 25)
```

```python
q3 = np.percentile(sorted_rscore, 75)
```

```python
print(q1,q3)
```

```python
IQR = q3-q1
```

```python
IQR
```

```python
lwr_bound = q1-(1.5*IQR) 
```

```python
upr_bound = q3+(1.5*IQR) 
```

```python
print(lwr_bound, upr_bound)
```

```python
r_outliers = [] 
```

```python
for i in sorted_rscore: 
 if (i<lwr_bound or i>upr_bound): 
  r_outliers.append(i) 
print(r_outliers)
```
